```python
import numpy as np
import matplotlib.pyplot as plt
import seatease as sb
```

```python
# Lists available devices
sb.cseatease.SeaTeaseAPI().list_devices()
```

```python
# Load first device
dev = sb.cseatease.SeaTeaseAPI().list_devices()[0]
print(dev)
```

```python
# Using .f functionality
plt.figure()
for it in [3*1000,10*1000,100*1000,1000*1000]:
    dev.f.spectrometer.set_integration_time_micros(it)
    plt.plot(dev.f.spectrometer.get_wavelengths(),
             dev.f.spectrometer.get_intensities(),
             label=it//1000)
plt.xlim(400,600)
plt.xlabel(r"Wavelength (nm)")
plt.ylabel(r"Counts (arb. units)")
plt.legend(title=r"IT (ms))")
plt.show()
```

```python
# Using .features functionality
plt.figure()
for it in [3*1000,10*1000,100*1000,1000*1000]:
    dev.features["spectrometer"][0].set_integration_time_micros(it)
    plt.plot(dev.features["spectrometer"][0].get_wavelengths(),
             dev.features["spectrometer"][0].get_intensities(),
             label=it//1000)
plt.xlim(400,600)
plt.xlabel(r"Wavelength (nm)")
plt.ylabel(r"Counts (arb. units)")
plt.legend(title=r"IT (ms))")
plt.show()
```
